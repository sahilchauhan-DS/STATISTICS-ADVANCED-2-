1. What is hypothesis testing in statistics?
a. Hypothesis testing is a method for making decisions or inferences about population parameters based on sample data.
b. It involves formulating two hypotheses and using statistical evidence to accept or reject one of them.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
a. The null hypothesis (H‚ÇÄ) represents the default or no-effect assumption (e.g., no difference or no relationship).
b. The alternative hypothesis (H‚ÇÅ or Ha) represents what you aim to support ‚Äì a change, difference, or effect.
c. Hypothesis testing evaluates whether data provides sufficient evidence to reject H‚ÇÄ in favor of H‚ÇÅ.

3. What is the significance level in hypothesis testing, and why is it important?
a. The significance level (Œ±) is the threshold for rejecting the null hypothesis (commonly 0.05).
b. It represents the probability of making a Type I error ‚Äì rejecting a true null hypothesis.
c. It controls the false positive rate in hypothesis testing.

4. What does a P-value represent in hypothesis testing?
a. The P-value is the probability of obtaining a test statistic as extreme as or more extreme than the observed result, assuming the null hypothesis is true.
b. It measures the strength of evidence against the null hypothesis.

5. How do you interpret the P-value in hypothesis testing?
a. If the P-value ‚â§ Œ± (e.g., 0.05), reject the null hypothesis ‚Äì the result is statistically significant.
b. If the P-value > Œ±, fail to reject the null hypothesis ‚Äì insufficient evidence to support the alternative.

6. What are Type 1 and Type 2 errors in hypothesis testing?
a. Type I error: Rejecting a true null hypothesis (false positive).
b. Type II error: Failing to reject a false null hypothesis (false negative).
c. There is a trade-off between the two; reducing one often increases the other.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
a. A one-tailed test checks for a deviation in only one direction (e.g., mean > value).
b. A two-tailed test checks for deviations in both directions (e.g., mean ‚â† value).
c. The choice depends on the research question or hypothesis.

8. What is the Z-test, and when is it used in hypothesis testing?
a. The Z-test is used to test hypotheses about population means or proportions when the population standard deviation is known.
b. It assumes normal distribution or large sample sizes (n ‚â• 30).

10. What is the T-distribution, and when should it be used instead of the normal distribution?
a. The T-distribution is used when the sample size is small (typically n < 30) and the population standard deviation is unknown.
b. It is wider and has heavier tails than the normal distribution.
c. As sample size increases, it approaches the normal distribution.

11. What is the difference between a Z-test and a T-test?
a. Z-test: Used when population standard deviation is known and/or sample size is large.
b. T-test: Used when population standard deviation is unknown and sample size is small.
c. Both test means but under different conditions.

12. What is the T-test, and how is it used in hypothesis testing?
a. A T-test assesses whether the means of one or more groups differ significantly from each other or a known value.
b. Types include one-sample, two-sample (independent), and paired t-tests.
c. It uses the t-distribution to determine significance.

13. What is the relationship between Z-test and T-test in hypothesis testing?
a. Both tests evaluate differences between means.
b. The T-test is a generalization of the Z-test when the population variance is unknown.
c. When sample size is large, both give similar results.

14. What is a confidence interval, and how is it used to interpret statistical results?
a. A confidence interval gives a range of plausible values for a population parameter.
b. It reflects the uncertainty around a point estimate.
c. If the interval includes the null value (e.g., 0 for differences), the result may not be statistically significant.

15. What is the margin of error, and how does it affect the confidence interval?
a. The margin of error defines how much the estimate might vary due to sampling error.
b. It determines the width of the confidence interval.
c. Larger sample sizes reduce the margin of error.

16. How is Bayes' Theorem used in statistics, and what is its significance?
a. Bayes' Theorem calculates the probability of a hypothesis given observed data.
b. It combines prior beliefs with new evidence (posterior probability).
c. Widely used in decision making, machine learning, and diagnostics.

17. What is the Chi-square distribution, and when is it used?
a. The Chi-square distribution is a right-skewed distribution used in categorical data analysis.
b. It‚Äôs applied in tests of independence, homogeneity, and goodness of fit.
c. Based on the sum of squared standard normal variables.

18. What is the Chi-square goodness of fit test, and how is it applied?
a. It tests whether observed frequencies differ from expected frequencies.
b. Common in assessing how well data fits a theoretical distribution.
c. Requires categorical data and sufficiently large sample sizes.

19. What is the F-distribution, and when is it used in hypothesis testing?
a. The F-distribution is used to compare two variances.
b. It is asymmetric and skewed to the right.
c. It‚Äôs used in ANOVA and regression model comparisons.

20. What is an ANOVA test, and what are its assumptions?
a. ANOVA (Analysis of Variance) compares the means of three or more groups.
b. Assumptions:
‚ÄÉ‚ÄÉi. Independence of observations
‚ÄÉ‚ÄÉii. Normality of the residuals
‚ÄÉ‚ÄÉiii. Homogeneity of variances

21. What are the different types of ANOVA tests?
a. One-way ANOVA ‚Äì compares means across one factor.
b. Two-way ANOVA ‚Äì examines interaction between two factors.
c. Repeated measures ANOVA ‚Äì used when the same subjects are measured multiple times.

22. What is the F-test, and how does it relate to hypothesis testing?
a. The F-test evaluates whether two population variances are significantly different.
b. It is used in ANOVA and regression analysis.
c. A large F-value suggests that the group means are not all equal.

```python
# 15. Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm

sample = np.random.normal(loc=52, scale=10, size=50)
population_mean = 50
sample_mean = np.mean(sample)
std_dev = np.std(sample, ddof=1)
n = len(sample)

z = (sample_mean - population_mean) / (std_dev / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"Z-statistic: {z:.2f}")
print(f"P-value: {p_value:.4f}")
```

```python
# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = np.random.normal(102, 10, 50)
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
pop_mean = 100

z = (sample_mean - pop_mean) / (sample_std / np.sqrt(len(data)))
p_value = 2 * (1 - norm.cdf(abs(z)))

# Visualization
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)
critical_z = norm.ppf(0.975)

plt.plot(x, y, label='Standard Normal Distribution')
plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
plt.axvline(critical_z, color='green', linestyle=':', label='Critical Value (¬±1.96)')
plt.axvline(-critical_z, color='green', linestyle=':')
plt.fill_between(x, y, where=(x <= -critical_z) | (x >= critical_z), color='lightgray', alpha=0.5, label='Rejection Region')
plt.legend()
plt.title("Two-Tailed Z-test Decision Region")
plt.grid(True)
plt.show()
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05):
    se = sigma / np.sqrt(n)
    z_crit = norm.ppf(1 - alpha)
    x = np.linspace(90, 115, 1000)

    plt.plot(x, norm.pdf(x, mu0, se), label='H0 Distribution', color='blue')
    plt.plot(x, norm.pdf(x, mu1, se), label='H1 Distribution', color='orange')

    critical_value = mu0 + z_crit * se
    plt.axvline(critical_value, color='red', linestyle='--', label='Critical Value')

    plt.fill_between(x, 0, norm.pdf(x, mu0, se), where=(x >= critical_value), color='red', alpha=0.3, label='Type I Error (Œ±)')
    plt.fill_between(x, 0, norm.pdf(x, mu1, se), where=(x < critical_value), color='purple', alpha=0.3, label='Type II Error (Œ≤)')

    plt.legend()
    plt.title("Type I and Type II Errors Visualization")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.grid(True)
    plt.show()

visualize_errors()
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance
import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(len(data))
    z = norm.ppf((1 + confidence) / 2)
    return mean - z * std_err, mean + z * std_err

sample = np.random.normal(100, 10, 50)
ci = confidence_interval(sample)
print(f"95% Confidence Interval: {ci}")
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

# Bayes' Theorem: P(A|B) = (P(B|A) * P(A)) / P(B)

def bayes_theorem(p_a, p_b_given_a, p_b_given_not_a):
    # P(not A) = 1 - P(A)
    p_not_a = 1 - p_a
    # Total probability of B (P(B)) using the law of total probability
    p_b = (p_b_given_a * p_a) + (p_b_given_not_a * p_not_a)
    # Applying Bayes' Theorem to find P(A|B)
    p_a_given_b = (p_b_given_a * p_a) / p_b
    return p_a_given_b

# Example: A medical test for a rare disease
# P(Disease) = 0.01 (1% of population has the disease)
# P(Pos | Disease) = 0.99 (Test is 99% accurate for people with disease)
# P(Pos | No Disease) = 0.05 (5% false positive rate)

p_disease = 0.01  # Prior probability of disease
p_positive_given_disease = 0.99  # Likelihood of positive test result if the person has the disease
p_positive_given_no_disease = 0.05  # Likelihood of positive test result if the person doesn't have the disease

posterior = bayes_theorem(p_disease, p_positive_given_disease, p_positive_given_no_disease)

print(f"Probability of having disease given positive test result: {posterior:.4f}")
```

```python
# 12. Perform a Chi-square test for independence between two categorical variables in Python

import numpy as np
from scipy.stats import chi2_contingency

# Observed data for two categorical variables (e.g., gender and choice of product)
# The data is represented as a 2x2 contingency table
observed = np.array([[30, 10],   # Gender 1 (30 choose product A, 10 choose product B)
                     [15, 25]])  # Gender 2 (15 choose product A, 25 choose product B)

# Perform the chi-square test of independence
chi2_stat, p_val, dof, expected = chi2_contingency(observed)

print(f"Chi-squared statistic: {chi2_stat:.2f}")
print(f"P-value: {p_val:.4f}")
print(f"Degrees of freedom: {dof}")
print(f"Expected frequencies: \n{expected}")
```
